```python
import camb
import numpy as np
import matplotlib.pyplot as plt
```

```python
pars = camb.read_ini('https://raw.githubusercontent.com/cmbant/CAMB/master/inifiles/planck_2018.ini')
results = camb.get_results(pars)
cl = results.get_total_cls(CMB_unit='muK')
ls = np.arange(cl.shape[0])
cl_r0 = np.array([cl[:,i] for i in range(4)])
```

```python
for i in range(4):
    plt.plot(cl_r0[i])
plt.loglog()
```

```python
#Add tensors
pars.WantTensors=True
pars.InitPower.set_params(r=1)
results = camb.get_results(pars)
cl = results.get_total_cls(CMB_unit='muK')
cl_r1 = np.array([cl[:,i] for i in range(4)])
```

```python
for i in range(3):
    plt.plot(cl_r0[i])
    plt.plot(cl_r1[i])
plt.loglog()
```
